```python
data_only_species = pd.read_csv('/Users/milenabiernacka/Desktop/praca-magisterska/DF/DF20M-metadata/DF20M-train_metadata_PROD.csv')
data_with_poisonous = pd.read_csv('/Users/milenabiernacka/Desktop/praca-magisterska/DF/validation/FungiCLEF2023_val_metadata_PRODUCTION.csv')
```

```python
# Usunięcie duplikatów z 'data_with_poisonous' przy założeniu, że dla danego gatunku kolumna 'poisonous' powinna być jednolita
unique_poisonous = data_with_poisonous.drop_duplicates(subset='species', keep='first')

# Teraz możemy przeprowadzić bezpieczne scalenie (merge), które nie spowoduje nadmiarowych wierszy
merged_df = pd.merge(data_only_species, unique_poisonous[['species', 'poisonous']], on='species', how='inner')

# 'merged_df' zawiera teraz unikalne wiersze z 'data_only_species' z dodaną informacją o 'poisonous' z 'data_with_poisonous'
merged_df.head()  
```

```python
merged_df
```

```python
# Sprawdzanie, czy dla każdego gatunku w kolumnie 'species' wartości w kolumnie 'poisonous' są zawsze takie same
consistent_species = merged_df.groupby('species').nunique()['poisonous'] == 1

# Wypisanie gatunków, które mają różne wartości w kolumnie 'poisonous'
inconsistent_species = consistent_species[consistent_species == False].index.tolist()

# Wynik: Lista gatunków, które mają niejednoznaczne wartości w kolumnie 'poisonous'
inconsistent_species
```
